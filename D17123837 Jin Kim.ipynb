```r
library(ggplot2)
library(lubridate)
library(sqldf)
library(tidyverse)
library(scales)
library(gridExtra)
```

```r
#covid19 : https://data.europa.eu/euodp/en/data/dataset/covid-19-coronavirus-data/resource/260bbbde-2316-40eb-aec3-7cd7bfc2f590
#amazon : https://finance.yahoo.com/quote/AMZN/history?p=AMZN
#netflix : https://finance.yahoo.com/quote/NFLX/history?p=NFLX
#apple : https://finance.yahoo.com/quote/NFLX/history?p=AAPPL

covid19 <- read.csv(file = "Covid-19.csv")
amazon <- read.csv(file = "Amazon.csv")
netflix <- read.csv(file = "Netflix.csv")
apple <- read.csv(file = "Apple.csv")
names(covid19)[names(covid19)=="dateRep"]<-"Date"
covid19$Date <- as.Date(covid19$Date, format="%d/%m/%y")
```

```r
# The number of cases that were infected by covid-19 in each month between January and April 2020

covid_case_df_jan = sqldf("select day, sum(cases) as 'number of cases' 
                        from covid19
                        where month=1
                        group by day 
                        having count(cases) > 1")

covid_case_df_feb = sqldf("select day, sum(cases) as 'number of cases' 
                        from covid19
                        where month=2
                        group by day 
                        having count(cases) > 1")

covid_case_df_mar = sqldf("select day, sum(cases) as 'number of cases' 
                        from covid19
                        where month=3
                        group by day 
                        having count(cases) > 1")

covid_case_df_apr = sqldf("select day, sum(cases) as 'number of cases' 
                        from covid19
                        where month=4
                        group by day 
                        having count(cases) > 1")

plot_cases<-ggplot(covid19, aes(x=day, y= max('number of cases'))) +  coord_cartesian(ylim=c(0,max(covid_case_df_apr$'number of cases')))+
            geom_line(data=covid_case_df_jan,aes(color='January', y = covid_case_df_jan$'number of cases'), lwd=1)+
            geom_line(data=covid_case_df_feb,aes(color='February', y = covid_case_df_feb$'number of cases'), lwd=1)+
            geom_line(data=covid_case_df_mar,aes(color='March', y = covid_case_df_mar$'number of cases'), lwd=1)+
            geom_line(data=covid_case_df_apr,aes(color='April', y = covid_case_df_apr$'number of cases'), lwd=1)+
            labs(x='Day', y='Number of Cases', title='Number of Cases by month')
plot_cases 
```

```r
#change name of attributes to merge datasets
# Date : Date that the market opens
# Open : The price of share when the market opened
# High : The highest value the share reached on that day
# Low : The lowest value the share reached on that day
# Close : The price of share when the market closed
# Adj.Close : The adjusted closing price shows the stock's value after posting a dividend
# Volume : The number of shares traded on that day

names(amazon)[names(amazon)=="Open"]<-"amazon_Open"
names(amazon)[names(amazon)=="High"]<-"amazon_High"
names(amazon)[names(amazon)=="Low"]<-"amazon_Low"
names(amazon)[names(amazon)=="Close"]<-"amazon_Close"
names(amazon)[names(amazon)=="Adj.Close"]<-"amazon_Adj.Close"
names(amazon)[names(amazon)=="Volume"]<-"amazon_Volume"

names(netflix)[names(netflix)=="Open"]<-"netflix_Open"
names(netflix)[names(netflix)=="High"]<-"netflix_High"
names(netflix)[names(netflix)=="Low"]<-"netflix_Low"
names(netflix)[names(netflix)=="Close"]<-"netflix_Close"
names(netflix)[names(netflix)=="Adj.Close"]<-"netflix_Adj.Close"
names(netflix)[names(netflix)=="Volume"]<-"netflix_Volume"

names(apple)[names(apple)=="Open"]<-"apple_Open"
names(apple)[names(apple)=="High"]<-"apple_High"
names(apple)[names(apple)=="Low"]<-"apple_Low"
names(apple)[names(apple)=="Close"]<-"apple_Close"
names(apple)[names(apple)=="Adj.Close"]<-"apple_Adj.Close"
names(apple)[names(apple)=="Volume"]<-"apple_Volume"
```

```r
# The number of deaths that were infected by covid-19 in each month between January and April 2020


covid_death_df_jan = sqldf("select day, sum(deaths) as 'number of deaths' 
                        from covid19
                        where month=1
                        group by day 
                        having count(deaths) > 1")

covid_death_df_feb = sqldf("select day, sum(deaths) as 'number of deaths' 
                        from covid19
                        where month=2
                        group by day 
                        having count(deaths) > 1")

covid_death_df_mar = sqldf("select day, sum(deaths) as 'number of deaths' 
                        from covid19
                        where month=3
                        group by day 
                        having count(deaths) > 1")

covid_death_df_apr = sqldf("select day, sum(deaths) as 'number of deaths' 
                        from covid19
                        where month=4
                        group by day 
                        having count(deaths) > 1")

plot_deaths<-ggplot(covid19, aes(x=day, y= max('number of deaths'))) +  coord_cartesian(ylim=c(0,max(covid_death_df_apr$'number of deaths')))+
            geom_line(data=covid_death_df_jan,aes(color='January', y = covid_death_df_jan$'number of deaths'), lwd=1)+
            geom_line(data=covid_death_df_feb,aes(color='February', y = covid_death_df_feb$'number of deaths'), lwd=1)+
            geom_line(data=covid_death_df_mar,aes(color='March', y = covid_death_df_mar$'number of deaths'), lwd=1)+
            geom_line(data=covid_death_df_apr,aes(color='April', y = covid_death_df_apr$'number of deaths'), lwd=1)+
            labs(x='Day', y='Number of Deaths', title='Number of Deaths by month')
plot_deaths 
```

```r
#change name of attributes to merge datasets

# Date : Date that the market opens
# Open : The price of share when the market opened
# High : The highest value the share reached on that day
# Low : The lowest value the share reached on that day
# Close : The price of share when the market closed
# Adj.Close : The adjusted closing price shows the stock's value after posting a dividend
# Volume : The number of shares traded on that day

names(amazon)[names(amazon)=="Open"]<-"amazon_Open"
names(amazon)[names(amazon)=="High"]<-"amazon_High"
names(amazon)[names(amazon)=="Low"]<-"amazon_Low"
names(amazon)[names(amazon)=="Close"]<-"amazon_Close"
names(amazon)[names(amazon)=="Adj.Close"]<-"amazon_Adj.Close"
names(amazon)[names(amazon)=="Volume"]<-"amazon_Volume"

names(netflix)[names(netflix)=="Open"]<-"netflix_Open"
names(netflix)[names(netflix)=="High"]<-"netflix_High"
names(netflix)[names(netflix)=="Low"]<-"netflix_Low"
names(netflix)[names(netflix)=="Close"]<-"netflix_Close"
names(netflix)[names(netflix)=="Adj.Close"]<-"netflix_Adj.Close"
names(netflix)[names(netflix)=="Volume"]<-"netflix_Volume"

names(apple)[names(apple)=="Open"]<-"apple_Open"
names(apple)[names(apple)=="High"]<-"apple_High"
names(apple)[names(apple)=="Low"]<-"apple_Low"
names(apple)[names(apple)=="Close"]<-"apple_Close"
names(apple)[names(apple)=="Adj.Close"]<-"apple_Adj.Close"
names(apple)[names(apple)=="Volume"]<-"apple_Volume"
```

```r
amazon_netflix_stock <- merge(amazon, netflix, by="Date")
total_stock <- merge(amazon_netflix_stock, apple, by="Date")

# amazon stock
# the graph shows that amazon stock price plummeted when the number of covid-19 cases has significantly risen
# the back line represnets that lowest price and highest price of that day.
# the front line represents the open price of the day
amazon_open = sqldf("select Date, amazon_Open as 'Open price in stock', amazon_Low, amazon_High from total_stock where Date between '2020-01-01' and '2020-04-30'")

amazon_plot <- ggplot(amazon_open, aes(x=as.Date(Date), y=`Open price in stock`)) + 
               geom_ribbon(aes(ymin=amazon_Low, ymax=amazon_High), alpha=0.2) +
               geom_line(colour='red', lwd=1)+
               labs(x='Date', y='Stock Price in Amazon', title = 'Amazon Stock in 2020' )

amazon_plot
```

```r
# netflix stock
# the graph shows that amazon stock price plummeted when the number of covid-19 cases has significantly gone up
# the back line represnets that lowest price and highest price of that day.
# the front line represents the open price of the day
netflix_open = sqldf("select Date, netflix_Open as 'Open price in stock', netflix_Low, netflix_High from total_stock where Date between '2020-01-01' and '2020-04-30'")
            
netflix_plot <- ggplot(netflix_open, aes(x=as.Date(Date), y=`Open price in stock`)) +
                geom_ribbon(aes(ymin=netflix_Low, ymax=netflix_High), alpha=0.2) +
                geom_line(colour='blue', lwd=1)+
                labs(x='Date', y='Stock Price in Netflix', title='Netflix Stock in 2020')
    
netflix_plot
```

```r
# apple stock
# the graph shows that amazon stock price plummeted when the number of covid-19 cases has significantly increased
# the back line represnets that lowest price and highest price of that day.
# the front line represents the open price of the day
apple_open = sqldf("select Date, apple_Low, apple_High, apple_Open as 'Open price in stock' from total_stock where Date between '2020-01-01' and '2020-04-30'")

apple_plot <- ggplot(apple_open, aes(x=as.Date(Date), y=`Open price in stock`)) +
              geom_ribbon(aes(ymin=apple_Low, ymax=apple_High), alpha=0.2) +
              geom_line(colour='green', lwd=1)+
              labs(x='Date', y='Stock Price in Apple', title='Apple Stock in 2020')
apple_plot
```

```r
#Shows open price of each market
amazon_stock = sqldf("select Date, amazon_Open as 'Open price in stock' from total_stock where Date between '2020-01-01' and '2020-04-30'")
netflix_stock = sqldf("select Date, netflix_Open as 'Open price in stock' from total_stock where Date between '2020-01-01' and '2020-04-30'")
apple_stock = sqldf("select Date, apple_Open as 'Open price in stock' from total_stock where Date between '2020-01-01' and '2020-04-30'")

stock_plot <- ggplot(NULL, aes(x=as.Date(Date), y=`Open price in stock`)) +
              geom_line(data=amazon_stock, aes(colour="Amazon"), lwd=1.5) + 
              geom_line(data=netflix_stock, aes(colour="Netflix"), lwd=1.5)+
              geom_line(data=apple_stock, aes(colour="Apple"), lwd=1.5)+
              labs(x='Date', y='Stock Price', title='Stock Market in 2020')

stock_plot

```

```r
# Final Result
# All three companies were affected by coronavirus in the middle of march.

# The 4th graph shows that the number of cases has gone up siginificantly in March.

# The 5th graph shows the number of death from coronavirus, it is hard to judge that the stock market was affected by the number of deaths
# since the highest number of deaths is in April and the stock market still has gone up.

# The 6th graph presents that the companies have been recovering their industry gradually since they had the lowest stock price.

grid.arrange(amazon_plot,netflix_plot,apple_plot,ncol=3, nrow=1)
grid.arrange(plot_cases, plot_deaths, stock_plot, ncol=3, nrow=1)
```
